## 1.1

Selling a call option means that you will provide someone else with the decision on whether or not to buy stock from you at a fixed price; i.e., the buyer of your call option decides whether or not you will be compelled to sell stock at the agreed upon price.  Buying a put option means that you get to exercise the option to sell stock at an agreed upon price to another party.  The sold call will only be executed by the buyer when the strike price is lower than the market price, so that the buyer of the call can buy stock from you at a discount and sell it for a profit in the market.  The bought put will only be executed by you when the strike price is higher than the market, so that you can buy stock in the market and sell it for a profit through the put.  Profiting by selling a call requires the price of the option to exceed the amount you expect the stock to drop during the lifetime of the option.  Profiting by buying a put requires the underlying to drop in value by an amount exceeding the price of the option.

## 1.17

Options and future are zero-sum transactions as one side wins what the other side loses.  The value in an option (or future) is the difference in the strike price on the underlying compared to the actual price in the future.  Bond investors get coupons and principal; stock investors get dividends.  But options and futures traders--notably, _not_ investors--simply get or lose the difference in value to the underlying.

Of course, this is only true in isolation within the transaction.  Both parties obviously benefit as either speculators or hedgers, otherwise they wouldn't participate in the transaction.

## 1.18

A long forward contract on an underlying asset coupled with an equivalent European put option, strike price set to the forward price, provides a release valve to back out of an unfavorable deal.  If the asset is worth less than the forward price, the put option can be exercised to get out of the transaction all together.  The European nature of the put doesn't take away, as this portfolio will only have accesss to this asset at maturity anyway.  

If the deal is favorable, then the put can be left unexercised and the asset can be sold for profit on the difference between the actual price at maturity and the forward price.  The only downside to this transaction is the cost of the put option in case the underlying asset is worth less than the forward (and strike) price.
